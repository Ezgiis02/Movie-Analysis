```python
def replace(title):
    a=''
    for i in title:
        if i in '0123456789':
            a+=i
        else:
            a+=''
    if a=='':
        a='0'
    return a
```

```python
budget_check=tab['budget']
```

```python
budget_check=budget_check.apply(replace)
```

```python
budget_check.apply( lambda x: int(x))
```

```python
budget_check.apply( lambda x: int(x)).mean()
```

```python
budget_check.apply( lambda x: int(x)).min()
```
